# Transposed Convolution
기본적인 convolution은 주어진 Input Data에서 Feature Extraction해서 더 작은 형태로 Feature map을 만든다.

이때 Deconvolution이라는 개념은 Convolution의 역연산으로 다시 Feature map의 크기를 확대하는데 사용한다.

이때 주어진 Featemap 사이사이에 padding을 넣는 방법을 기본적으로 Deconv라고 한다.



# 근데 이와는 다르게 transposed conv가 사용이 가능하다. 
transposed Convolution을 통해서 우리는 역연산이 가능하다.
여기서 transpose라는 말은 기존 convolution에서 사용된 kernel을 transpose만 하면, 다시 input size를 가질 수 있다는 말에서 나왔다.
(위에 나온 기본적인 Deconvolution은 feature_map 사이사이에 패딩을 넣어주는 방식이라면, transposed는 그냥 weight만 transpose시키면 된다.)

# Deconvolution
Input과 output을 바꾸고, 기존에 있던 weight matrix(kernel)을 transpose시키기만 하면 됌



```python
import torch
import torch.nn as nn
```

```python
# 일반적인 conv
input = torch.randn(1,16,12,12)
downsample = nn.Conv2d(16,16,3, stride = 2, padding = 1) # 16*16*3*3
h = downsample(input)
h.size()
```

```python
upsample = nn.ConvTranspose2d(16,16,3, stride=2, padding =1) # 16*16*3*3
predict = upsample(h,output_size = input.size())
predict.shape
```

# 정리하면

0) Deconvolution은 upsampling 기법이다.
1) 일반적인 Deconvolution은 주어진 input 사이사이에 padding을 집어넣는 방식으로 사용된다. <br>
2) Transpoed를 Deconvolution의 일종으로도 보지만, 주어진 메트릭스를 transposed 하는 방식으로 설명하기 때문에, 일반적인 Deconvolution과는 다르다고 할 수 있다. <br>
3) 이외에도 kernel 사이사이에 padding을 넣는 atrous convolution이 있기도 하다 -> 다양한 receptive field를 보는 피라미드 풀링과 같은 역할을 수행하기 도함. <br>
